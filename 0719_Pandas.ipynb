```python
# 일단 pandas가 있는지 확인
import numpy as np
import pandas as pd
```

```python
# series 만들자
# python의 list를 이용해서 만들자
s = pd.Series([-1,5,10,99], dtype = np.float64)  # Series 는 대문자
print(s)

print(s.values) # 현재 Series가 가지고 있는 값들을 ndarray로 가져옴 [-1,5,10,99] 결과

print(s.index) # 현재 Series가 가지고 index 값
                # RangeIndex(start=0, stop=4, step=1) 범위 형태로 표현한다.
                # 별도의 데이터 타입(Pandas가 제공)
```

```python
# Series는 index를 지정가능 (특징)

s = pd.Series([-1,5,10,99], dtype=np.float64,
             index=['c','b','a','k'])

print(s) # 숫자 index를 자기가 지정이 가능하다.
print(s['a']) # 문자열로 index 출력
print(s[2]) # 10.0 기존의 숫자 인덱스도 사용 할 수 있다. 0 1 2 3 - c b a k 로 치환 2 - b 이니 출력 10

# print(s['kk']) # 오류 index값이없는데 출력
```

```python
# 이번에는 index를 숫자로 다시 설정

s = pd.Series([-1,5,10,99], dtype=np.float64,
             index=[0,2,100,6])
print(s)

print(s[0]) # -1 출력
print(s[100]) # 10 출력

# print(s[1]) # 오류 설정, 숫자 index 설정 할 경우 기존의 남은 0 1 2 3가 없다. 그래서 순서 0 2 100 6 으로 사용


```

```python
# index 새로생성 할 경우 중복된 index 확인가능 ?

s = pd.Series([-1,5,10,99], dtype=np.float64,
             index=['c','b','c','k'])
print(s) # 중복된 index 사용 가능 c c

print(s['c']) # -1, 10 둘다 가지고 온다. 둘다출력
```

```python
import numpy as np
import pandas as pd
from datetime import date, datetime, timedelta #날짜 , 시간, 날짜 간격

start_day = datetime(2022,1,1)
print(start_day)

# 자동으로 규칙성 있는 데이터 자동 만들기
# list Comprehension
print([int(x) for x in np.random.normal(50,5,(10,))]) #평균 50 5인 난수 10개 추출

s1 = pd.Series([int(x)for x in np.random.normal(50,5,(10,))],
               index=[start_day + timedelta(days=x) for x in range(10)])
print(s1) # 하나씩 뜯어서 봐야한다.
```

```python
data = {'이름': ['홍길동','신사임당','강감찬','이순신','을지문덕'],
       '학과':['컴퓨터','철학','기계','건축','컴퓨터'],
       '학년':[1,2,2,4,3],
        '학점':[1.5, 2.0, 3.1, 1.1, 2.5 ]
       }
df = pd.DataFrame(data,
                 columns=['학과','이름','학점','학년','등급'],
                  # 가능하다. 단지 nan 값을 넣는다.
                 index=['one','two','three','four','five']
                 )
display(df)
#dataFrame 새로운 colum추가
# df['나이']=[20,21,22,22,20] # list => series
df['나이']=pd.Series([20,21,22,22,20]) # list => series 직접 이렇게 만들면
                                     # row index로 인해 값이 부여가 안됨
df['나이']=pd.Series([20,21,22,22,20],
                  index=['one','two','three','four','five']
                  )

display(df)
```

```python
# 4. dataFrame에서 colum의 추가 없는 column을 indexing 해서 값을 넣으며,
# 단 series를 추가하여 row index 추가

# 새로운 컴럼을 추가할 떄 ㄱ조의 컴러을 이용하여 테이터 추출
# df['장학벼'] = df['학점']

data = {'이름': ['홍길동','신사임당','강감찬','이순신','을지문덕'],
       '학과':['컴퓨터','철학','기계','건축','컴퓨터'],
       '학년':[1,2,2,4,3],
        '학점':[1.5, 2.0, 3.1, 1.1, 2.5 ]
       }
df = pd.DataFrame(data,
                 columns=['학과','이름','학점','학년','등급'],
                  # 가능하다. 단지 nan 값을 넣는다.
                 index=['one','two','three','four','five']
                 )
display(df)
# 불필요한 부분 삭제
df.drop('등급',axis=1,inplace=False)
display(df)
df.drop(['등급','학점'] ,axis=1, inplace=False)
display(df)
```
