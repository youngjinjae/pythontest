```python
# 또 흔하게 많이 사용되는 함수가 있는데 sort()
# 정렬하는거예요! DataFrame을 정렬하는 방법으로 사용되요!

import numpy as np
import pandas as pd

np.random.seed(100)  # 난수값을 고정시킬려는 목적

df = pd.DataFrame(np.random.randint(0,10,(6,4)))
display(df)

df.columns = ['A', 'B', 'C', 'D']
df.index = pd.date_range('20230101', periods=6)
display(df)

# index를 랜덤하게 섞을꺼예요!
# np.random.shuffle(df.index)  # Error! DataFrame의 index같은건 mutable하지 않아요!
# np.random.shuffle()은 인자로 들어온 놈을 랜덤하게 섞는 일을 수행해요!
# index는 특성 상 한번 정해지면 바꿀 수 없어요!
random_date = np.random.permutation(df.index)

print(random_date)

# 우리가 가지고 있는 DataFrame의 index를 바꿔줄꺼예요!
# DataFrame의 index가 없는 경우는 그냥 assign하면 되요!
# DataFrame의 index가 이미 존재하는 경우 다른 index로 바꿀려면 reindex()

df2 = df.reindex(index=random_date, columns=['B', 'A', 'D', 'C'])
display(df2)
```

```python
# 1. Missing Value부터 알아보도록 해요!
# 결치값인데 이거 어떻게 해야 하나요?
# 일단 결치값을 찾아서 삭제하거나 다른값으로 대체
# DataFrame에서 이런 결치값은 NaN으로 표현되요!
```

```python
# 예제는 타이타닉 데이터 셋으로 할꺼예요!
# 이 데이터셋은 어디서 구하나요? - Kaggle에서 구할 수 있구요
# seaborn이라는 module이 있는데 이 모듈이 데이터 셋을 제공해요!

import seaborn as sns

# seaborn이 제공하는 데이터셋이 있는데 그 중 타이타닉 데이터셋을 이용해 보아요!
df = sns.load_dataset('titanic')
# display(df)

print(df.shape)  # (891, 15)   총 891명이 있고 15개의 column이 있어요!
# print(df.info())

print(df['deck'].value_counts(dropna=False))  # NaN이 688개 있어요!!

# 결치값이 몇개 있는지 확인하는 또 다른 방법

display(df.isnull())  # 전체 DataFrame에 대한 mask를 얻을 수 있어요!

# DataFrame의 함수중에 sum(axis=0)함수가 있고 Python은 True가 1 False가 0으로
# 계산되기 때문에..

display(df.isnull().sum(axis=0))
```

```python
# 이렇게 결치값을 찾아냈으면 그 다음에는 처리해야 되요!
# 가장 쉽고 좋은 방식은 결치값이 있는 행을 혹은 열을 삭제하는 거예요
# 하지만 까뜩이나 데이터 수가 적은데 삭제하는건 사실 부담이 따라요!!

# 삭제해 볼꺼예요!
# dropna()라는 함수를 이용해서 결치값을 찾아 그 행이나 열을 삭제하는거예요!
new_df = df.dropna(axis=1)  # 결치값이 들어가 있는 행을 찾아서 삭제
display(new_df)  # 11개

# 그런데 결치값이 아주 조금이예요!상대적으로 전체 데이터 많아요!

new_df = df.dropna(axis=1, thresh=500)  # 결치값이 들어가 있는 행을 찾아서 삭제
#  thresh를 이용하며 결치값의 개수에 따라 삭제 여부가 결정되요!\

display(new_df)  # 891 rows × 14 columns
```

```python
# 결치값을 발견하면 삭제하거나 아니면 치환을 해야 해요!

# 우리 예제에서 승선도시를 나타내는게 embark_town인데
# 아까 조사를 해 보니 결치가 2개있어요!
# 이런경우에는 column을 통으로 삭제하는거 보다 다른 값으로 채우는게 더 좋을 거
# 같아요!

display(df['embark_town'].fillna(df['embark_town'].value_counts(dropna=False).idxmax(),
                        inplace=False))
```

```python
# 방금 살펴본건.. DataFrame의 연결!!
# 연결말고 결합도 있어요! (SQL의 Join과 똑같아요!)

data1 = {
    '학번': [1, 2, 3, 4],
    '이름': ['홍길동', '신사임당', '강감찬', '이순신'],
    '학년': [2, 4, 1, 3]
}

data2 = {
    '학번': [1, 2, 4, 5],
    '학과': ['컴퓨터', '수학과', '철학', '건축과'],
    '학점': [3.4, 1.2, 4.1, 3.5]
}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)
display(df1)
display(df2)

display(pd.merge(df1, df2, on='학번', how='inner'))  # iiner join
display(pd.merge(df1, df2, on='학번', how='outer'))  # outer join
display(pd.merge(df1, df2, on='학번', how='left'))  # left outer join
display(pd.merge(df1, df2, on='학번', how='right'))  # right outer join
```
