```python
import numpy as np
import pandas as pd

df = pd.read_csv('./data/mpg/auto-mpg.csv', header=None)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower',
              'weight', 'acceleration', 'model year', 'origin', 'name']
display(df)

# 몇개의 중요한 함수 사용
# count()
# dataframe이 가지고 있는 각 열의 데이터 개수를 series를 리턴
# 결치값 여부 확인 할 때 사용
print(df.count())
# value_counts()는 Series 대상으로 안에 있는 고유값의 개수를 알려줌
# origin (나라관련) = 자동차 제조국의 대한 코드 1- 미국, 2-유럽, 3- 일본

print(df['origin'].value_counts())

# 한가지 조심 count() 사용 시에는 Not a number=NaN 를 사용하지않음
# 2차원 ndarray를 만들고 여기에 column과 index를 붙여서 dataframe 만듬
# count() 적용
# 결치값을 표현하는 numpy상수는 np.nan
# data = [[2,NaN]]
data = [[2,np.nan],
        [7,-3],
        [np.nan,np.nan],
        [1,-2]]
df = pd.DataFrame(data,
                  columns=['one','two'],
                  index=['a','b','c','d'])
print(df.count())
```

```python
import numpy as np
import pandas as pd

np.random.seed(100) # 난수값 고정 시킬 목적
df = pd.DataFrame(np.random.randint(0,10,(6,4))) # 0으로 시작 10 까지 난수 중에 6행 4열 만들기
display(df)

df.columns = ['A','B','C','D']
df.index = pd.date_range('20230101',periods=6)
display(df) # pandas 날짜 범위 쉽게 표현 해준다. index 값을

# index를 랜덤하게 썩음
# np.random.shuffle(df.index) error dataframe의 index같은 것 mutable 못한다.
# np.random.shuffle()은 인자로 들어온 값을 랜덤하게 사용하는 수행함
# index는 특성 상 한번 정해지면 변경 불가능
random_date = np.random.permutation(df.index)

print(random_date)

# 우리가 가지고 있는 dataframe의 index 벼경
# dataframe의 index가 없는 경우 그냥 assign 사용
# dataframe의 index가 이미 존재하는 경우 index 바꿀
# reindex 사용

df2 = df.reindex(index=random_date, columns=['B','A','D','C'])
display(df2)
```

```python
# 우리가 원하는 dataFrame 생성
# sort 이용해서 dataframe 정렬
# sort 2가지 방법
# 하나는 index 정렬 하는 방법
# 또 다른 하나 value 정렬
# index 로 정렬

display(df2.sort_index(axis=1, ascending=True) )
# 위에 보면 B A D C 순서가 이상했는데
# 아래의 결과는 오름차순으로 정렬
display(df2.sort_index(axis=0, ascending=True) )
# 날짜 별로 오름차순 정렬

# 이번에는 value 값으로 정렬
display(df2.sort_values(by='B'))


```

```python
# 예제는 타이타닉 데이터 셋으로 예상
# 타이타닉 데이터셋 - Kaggle 에 데이터 set 구할 수 있다.
# seaborn이라는 module 있는데 이 모듈이 데이터 셋을 제공
# 다른 프롬포트 켜서 activate data_env
# conda install seaborn
import seaborn as sns

# seaborn이 제공하는 데이터셋이 있는데 그 중에 타이타닉 데이터셋 이용
df = sns.load_dataset('titanic')
display(df)
print(df.shape) #(891,15) 총 891명있고 15개의 column
print(df.info())
print(df['deck'].value_counts(dropna=False))
# NaN 688개 있음

display(df.head(3).isnull()) # 전체 dataframe에 대해서 mask 얻을수 있음
# DataFrame의 함수중에 sum(axis=0)

display(df.isnull().sum(axis=0))
```

```python
# 이렇게 결치값을 찾아내면 그다음에 처리
# 가장 쉽고 좋은 방식은 결치값이 행 또는 열 삭제
# 데이터수가 적은데 삭제하는 거은 부담

# 삭제
# dropna()라는 함수를 이용해서 결치값을 찾아 행이나 열 삭제
new_df = df.dropna(axis=1) # 결치밧이 들어가는 행을 찾아서 삭제 11로 줄어듬 15에서 11로
display(new_df)#891x11

# 그런데 결치값이 아주 조금이고, 상대적으로 데이터 많을 경우

new_df = df.dropna(axis=1,thresh=500) #결치값 들어가서 행 삭제
display(new_df) # 891x14  컬럼 하나 삭제
```

```python
# 결치값을 발견하는 삭제하거나 아니면 치환
# 우리 예제에서 승선도시에 나타내는 embarked_town 에서
# 아까 조사해보니 결치가 2개
# 이런 경우 column을 통으로 삭제하는거 보다 다른 값을 채운느게 좋은가?

df['embark_town'].value_counts(dropna=False)
# df['embark_town'].value_counts(dropna=False).idxmax() 최대 값 채우기
display(df['embark_town'].fillna(df['embark_town'].value_counts(dropna=False).idxmax()
                                 ,inplace=False)) # inplace는 , 이다. 기억
```

```python
# dataFrame자체를 연결하는 방법
df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': [1, 2, 3, 4],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3'],
                    'd': ['d0', 'd1', 'd2', 'd3'],
                    'e': ['e0', 'e1', 'e2', 'e3']},
                   index=[2, 3, 4, 5])

display(df1,df2)
# result1 = pd.concat([df1,df2],axis=0)
# display(result1) # 기존인덱스 보면 중복 숫자나옴

result1 = pd.concat([df1,df2],axis=0,ignore_index=True) # index 무시해라
display(result1)

result1 = pd.concat([df1,df2],axis=1) # index 무시해라
display(result1) #열로 맞춰서 정렬
```

```python
# 방금 살펴본 경우 dataframe 의 연결
# 연결 말고 결함도 있음 (sql의 join 같은 효과)

data1 = {
    '학번':[1,2,3,4],
    '이름':['홍길동','신사임당','강감찬','이순신'],
    '학년':[2,4,1,3]

}
data2 = {
    '학번':[1,2,4,5],
    '학과':['컴퓨터','수학과','철학','스포츠'],
    '학점':[3.4,1.2,4.1,3.5]

}

df1 = pd.DataFrame(data1)
df2 = pd.DataFrame(data2)

# display(df1)
# display(df2)

display(pd.merge(df1,df2,on='학번',how='inner')) # inner join 으로
display(pd.merge(df1,df2,on='학번',how='outer')) # outer join
display(pd.merge(df1,df2,on='학번',how='left')) # left join
display(pd.merge(df1,df2,on='학번',how='right')) # right join
```
