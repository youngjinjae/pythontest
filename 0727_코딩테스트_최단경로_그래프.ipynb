```python
# 간단한 다익스트라 알고리즘
import sys
input = sys.stdin.readline
INF = int(1e9) # 무한의 의미의 값으로 10억 설정

# 노드의 개수, 간선의 개수 입력받기
# n,m = map(int, input().split())

n,m = map(int, input().split())

# 시작하는 노드 버호 입력받기
start = int(input())
# 각 노드에 연결되어 있는 노드에 대한 정보를 담는 리스트 만들기
graph = [[]for i in range(n+1)]
# 방문한 적이 있는지 체크하는 목적의 리스트를 만들기
visited = [False]*(n+1)
# 최단 거리 테이블 모두 무한으로 초기화
distance = [INF]*(n+1)

# 모든간선 정보 입력받기
for _ in range(m):
  a,b,c = map(int,input().split())
  # a 번 노드에서 b노드로 가는 비용이라는 c라는 의미
  graph[a].append((b, c))

# 방문한지 않는 노드 중에서 가장 최단거리가 짤은 노드 번호 반환
def get_smallest_node():
  min_value = INF
  index = 0 # 가장 최단 거리가 짧은 노드(인덱스)
  for i in range(1, n+1):
    if distance[i] < min_value and not visited[i]:
      min_value = distance[i]
      index = i
  return index

def dijkstra(start):
  # 시작 노드에 대해서 초기화
  distance[start] = 0
  visited[start] = True
  for j in graph[start]:
    distance[j[0]] =j[1]
  #시작 노드를 제오ㅓㅣ한 전체 n-1개 의 노드에 대해 반복
  for i in range(n-1):
    # 현재 최단 거리 짧은 노드 방문처리
    now = get_smallest_node()
    visited[now] = True
    # 현재 노드와 연결된 다른 노드를 확인
    for j in graph[now]:
      cost = distance[now]+j[1]
      # 현재 노드를 거쳐서 다른 노드로 이동하는 거리가 더 짧은 겨우
      if cost <  distance[j[0]]:
        distance[j[0]]= cost

# 다익스트라 알고리즘 수행
dijkstra(start)

# 모든 노드로 가기 위한 최단 거리 출력
for i in range(1,n+1):
  # 도달할 수 없는 경우, 무한 출력
  if distance[i] == INF:
    print("INFINITY")
    # 도달할 수 있는 경우 거리 출력
  else:
    print(distance[i])










```

```python
# 전보 예시
import heapq
import sys

input = sys.stdin.readline
INF = int(1e9) # 무한 의미 10억 설정

# 노드의 개수 간선의 개수 시작노드
n, m, start = map(int, input().split())
# 각노드에 연결 노드에 대한 정보를 리스트 만들기
graph = [[]for i in range(n+1) ]
# 최단 거리 테이블 모두 무한으로 초기화
distance = [INF]*(n+1)

# 모든 간선 정보 입력
for _ in range(m):
  x,y,z = map(int, input().split())
  #x 번 노드 y 번 노드 가는 비용 z라는 의미
  graph[x].append((y,z))

def dijkstra(start):
  q = []
  # 시작 노드로 가기 위한 최단 경로 0으로 설정하여 큐 삽입
  heapq.heappush(q,(0,start))
  distance[start]=0
  while q:
    dist, now = heapq.heappop(q) # 가장 최단 거리 짧은 노드 정보
    if distance[now] < dist:
      continue
    # 현재 노드와 연결된 다른 인접한 노드 확인
    for i in graph[now]:
      cost = dist+i[1]
      # 현재 노드 거쳐서 다른 노드 이동 거리 짧은 경우
      if cost < distance[i[0]]:
        distance[i[0]] = cost
        heapq.heappush(q,(cost,i[0]))

# 다 익스트라 알고리즘 수행
dijkstra(start)

# 도달 할 수 있는 노드의 개수
count = 0

# 도달 할수 있는 노드 가장 멀리 있는 노드, 최단거리
max_distance = 0
for d in distance :
  # 도달 할수 있는 노드
  if d != INF :
    count +=1
    max_distance = max(max_distance, d)

# 시작 노드 제외해야 count -1 출력
print(count -1,max_distance)





```

```python
# 크루스칼 알고리즘 뭐야

def find_parent(parent, x):
  # 루트노드가 아닌 루트노를 찾을떄까지 재귀 호출
  if parent[x] !=x:
    parent[x] = find_parent(parent,parent[x])
  return parent[x]

# 두 원소가 속한 집합 합치기

def union_parent(parent,a,b):
  a = find_parent(parent,a)
  b = find_parent(parent,b)
  if a < b:
    parent[b]=a
  else :
    parent[a]=b

# 노드의 개수 간선의 개수 입력
v,e = map(int,input().split())
parent = [0] * (v+1) # 상위 테이블 추가

# 모든 간선을 담은 리스트와 최종 비용 담을 변수
edges = []
result = 0

# 상위 테이블에서 상위 자기 자신을 초기화
for i in range(1,v+1):
  parent[i]=i

# 모든 간선에 대한 정보 입력받기
for _ in range(e):
  a, b, cost = map(int,input().split())
  # 비용 순으로 정렬 하기 튜플첫 번째 원소 비용 설정
  edges.append((cost,a,b))

# 간선을 비용 순으로 정렬
edges.sort()

# 간선을 하나씩 확인
for edge in edges:
  cost,a,b= edge
  # 사이클 발생하지 않는 경우에만 집합
  if find_parent(parent,a ) != find_parent(parent,b):
    union_parent(parent,a,b)
    result += cost

print(result)
```

```python
# 위상 정렬 소스코드
from collections import deque

# 노드의 개수와 간선의 개수 입력
v,e = map(int,input().split())
# 모든 노드에 대한 진입 차수 0 으로 초기화
indegree = [0]*(v+1)
# 각노드에 연결된 간선정보 담기 위한 연결리스트 초기화
graph = [[]for i in range(v+1)]

# 방향 그래프의 모든 간선 정보 입력받기
for _ in range(e):
  a, b =map(int,input().split())
  graph[a].append(b) # 점 a에서 b로 이동
  # 진입차수 1 증가
  indegree[b] +=1

# 위상 정렬 함수

def topology_sort():
  result = [] # 알고리즘 수행 결과담을 리스트
  q = deque()

  # 처음 시작할때 진입차수 0 인 노드에서 큐 삽입
  for i in range(1, v+1):
    if indegree[i] == 0:
      q.append(i)
  # 큐가 빌때 까지 반복
  while q:
    # 큐 원소 꺼내기
    now = q.popleft()
    result.append(now)
    # 해당 원소와 연결된 노드 진입차수 1빼기
    for i in graph[now]:
      indegree[i] -=1
      #새롭게 진입차수가 0 이 되는 노드를 큐삽입
      if indegree[i]==0:
        q.append(i)

  #d위상 정렬 수행 결과 출력
  for i in result:
    print(i, end=' ')
topology_sort()
```

```python
# 도시 분할 계획
# 특정 원소 속한 집합 찾기
def find_parent(parent,x):
  # 루트 노드가 아니라면, 루트 노드 찾을때 재귀 호출
  if parent[x]!=x:
    parent[x] = find_parent(parent,parent[x])
  return parent[x]

# 두 원소 집합 합치기
def union_parent(parent,a,b):
  a = find_parent(parent,a)
  b = find_parent(parent,b)
  if a < b:
    parent[b] = a
  else:
    parent[a] = b

# 노드의 개수와 간선의 개수 입력받기

v,e = map(int,input().split())
parent = [0] * (v+1) # 상위 테이블 초기화

# 모든 간선을 담을 리스트 최종 비용을 담을 변수
edges=[]
result = 0
# 상위 테이블에서 상위 자기 자신을 초기화
for i in range(1,v+1):
  parent[i] = i

# 모든 간선에 대한 정보 ㅇㅂ력
for _ in range(e):
  a,b,cost=map(int,input().split())
  # 비용 순으로 정렬 하고 튜플의 첫번째 원소 비용 설정
  edges.append((cost,a,b))

# 간선을 비용순으로 정렬
edges.sort()
last = 0 # 최소 신장 트리에 포함 가장 비용 큰 간선
# 간선을 하나씩 확인하며
for edge in edges :
  cost,a,b = edge
  # 사이클 발생하지 않는 경우에만 집합 포함
  if find_parent(parent,a) != find_parent(parent,b):
    union_parent(parent,a,b)
    result += cost
    last = cost
print(result - last)


```

```python
# 커리큘럼
from collections import deque
import copy
# 노드 개수 입력받기
v = int(input())

# 모든 노드에 대하 진입차수는 0 초기화
indegree = [0]*(v+1)
# 각 노드에 연결된 간선 정보 담기 위한 연결 리스트 초기화
graph = [[]for i in range(v+1)]
# 각 강의 시간을 0 초기화
time = [0]*(v+1)

# 방향 그래프 모든 간선 정보 입력받기
for i in range(1,v+1):
  data = list(map(int,input().split()))
  time[i] = data[0] # 첫 번째 수를 시간 정보를 담고 있다.
  for x in data[1:-1]:
    indegree[i] +=1
    graph[x].append(i)

# 위상 정렬 함수
def topology_sort():
  result = copy.deepcopy(time)# 알고리즘 수행 결과 리스트
  q = deque()

  # 처음 시작할떄 진입차수 0인 노드 큐삽입
  for i in range(1,v+1):
    if indegree[i] == 0:
      q.append(i)

  #큐가 빌 때 까지 반복
while q:
  # 큐에 원소 꺼내기
  now = q.popleft()
  # 해당하는 원소 연결된 노드들의 진입차수 1빼기
  for i in graph[now]:
    result[i] = max(result[i],result[now]+time[i])
    if indegree[i] == 0:
      q.append(i)
  # 위상 정렬 수행 결과
  for i in range(1,v+1):
    print(result[i])

topology_sort()


```
