```python
#   import

import numpy as np
import pandas as pd
import tensorflow as tf
import matplotlib.pyplot as plt
import matplotlib.image as img

#  2  .
# matplotlib figure   2   
fig = plt.figure(figsize=(10,10))
 #          , 10 10

ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,2,2)
#   x,y   

ori_image = img.imread('/content/girl-teddy.jpg')
ax1.imshow(ori_image)
print(ori_image.shape) #(429, 640, 3)  3

# convolution     API    
#   4
#   

# ( , height, width, color) => (1,429, 640, 3) 4
input_image = ori_image.reshape((1,) + ori_image.shape) #((1,)+429, 640, 3)  

print(input_image.shape)  #(1, 429, 640, 3)  
input_image = input_image.astype(np.float32) # s    32bit  

#### conv 
channel_1_input_image = input_image[:,:,:,0:1] #sling
# filter  
# filter  => (filter height,filter width,filter channel,filter )
# (filter height =3 ,filter width =3 ,filter channel = 1 ,filter  =1 )
weight = np.array([[[[-1]],[[0]],[[1]],
                   [[-1]],[[0]],[[1]],
                   [[-1]],[[0]],[[1]]]])
#stride = 1
# padding = valid
conv2d = tf.nn.conv2d(channel_1_input_image,
                     weight,
                     strides=[1,1,1,1],
                     padding='VALID').numpy()
# Conv2d  . 4 (  
conv2d_result=conv2d[0,:,:,:]
ax2.imshow(conv2d_result)
fig.tight_layout()
plt.show()





```
